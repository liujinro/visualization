```python
# Install Matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python

np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100,50)

# Basic scatter type of data
fig, ax = plt.subplots(figsize=(5, 3)) # Where we want subplots go and (5,3) is for default
ax.scatter(x,y)
```

```python
# Bar Graph
fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x,y)
```

```python
# Barh
fig, ax = plt.subplots(figsize=(5, 3))
ax.barh(x,y)
```

```python
# Activity
fig, ax = plt.subplots(figsize=(5, 3))
# ax.hist(x,y) Correction:
ax.hist(y)
```

```python
## Plot Labels
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
#(Definition of title, ylabel,xlabel)
ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
```

```python
# How we change labels and what they look like
# Key, what clolor of the key use

font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
```

```python
#Change
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1, loc='left')
ax.set_ylabel('Total growth', fontdict = font2,loc='top')
ax.set_xlabel('Years since start', fontdict = font2)
```

```python
# Activity

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets
import pandas as pd

# Load dataset
iris = datasets.load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Create contour plot
plt.figure(figsize=(5,3))
sns.kdeplot(data=iris_df, x="sepal width (cm)", y="sepal length (cm)", fill=True)
plt.title('Contour Plot of Sepal Width vs. Sepal Length')
plt.xlabel('sepal_width')
plt.ylabel('sepal_length')
plt.show()
```
